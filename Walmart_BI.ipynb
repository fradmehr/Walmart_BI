```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# --- Assume data already loaded as pandas DataFrames ---
# walmart_fact
# walmart_date

# 1. Join fact and date tables
df = walmart_date.copy()

# 2. Aggregate weekly sales
agg = (
    df.groupby(["STORE_ID", "DEPT_ID", "ISHOLIDAY"], as_index=False)
      .agg(TOTAL_SALES=("WEEKLY_SALES", "sum"))
)

# 3. Aggregate again to store-level (optional but cleaner for x-axis)
store_level = (
    agg.groupby(["STORE_ID", "ISHOLIDAY"], as_index=False)
       .agg(TOTAL_SALES=("TOTAL_SALES", "sum"))
)

# Pivot so each store has two columns: holiday vs non-holiday
pivot = store_level.pivot_table(index="STORE_ID", columns="ISHOLIDAY", values="TOTAL_SALES", fill_value=0)


# Plot side-by-side bars
ax = pivot.plot(kind="bar", color={'FALSE':"blue", 'TRUE':"gray"}, figsize=(10,6))

# Format y-axis to show in millions with 'M'
formatter = FuncFormatter(lambda x, _: f'{int(x/1e6)}M')
ax.yaxis.set_major_formatter(formatter)

# Add labels: tallest bar + holiday bar
store_ids = pivot.index
for i, store_id in enumerate(store_ids):
    # get the two bars for this store
    bars = [p for p in ax.patches if p.get_x() >= i-0.5 and p.get_x() < i+0.5]
    if bars:
        # tallest bar
        tallest = max(bars, key=lambda b: b.get_height())
        ax.annotate(f'{tallest.get_height()/1e6:.1f}M',
                    (tallest.get_x() + tallest.get_width()/2., tallest.get_height()),
                    ha='center', va='bottom', fontsize=8)
        # holiday bar (gray) if different from tallest
        for b in bars:
            if b.get_facecolor() == (0.5,0.5,0.5,1.0):  # gray RGBA
                ax.annotate(f'{b.get_height()/1e6:.1f}M',
                            (b.get_x() + b.get_width()/2., b.get_height()),
                            ha='center', va='bottom', fontsize=8)

plt.xticks(rotation=0)
plt.xlabel("Store ID")
plt.ylabel("Total Sales")
plt.title("Total Sales by Store and Holiday")
plt.legend(title="Is Holiday")
plt.tight_layout()
plt.show()


```
